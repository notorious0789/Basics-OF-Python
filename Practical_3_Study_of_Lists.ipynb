```python
# PROBLEM STATEMENT 2 (as Slicing one is first)
marks = [78, 85, 90, 66, 88]
# Find Maximum Marks:
print(max(marks))
# Find Minimum Marks:
print(min(marks))
# Find Total Marks:
print(sum(marks))
# Find Number of Students:
print(len(marks))
# Find Average Marks:
print(sum(marks)/len(marks))
# Sort Marks in ascending order
marks.sort()
print(marks)
# Sort Marks in descending order
marks.sort(reverse=True)
print(marks)
```

```python
# PROBLEM STATEMENT 3
''' 3. Grocery Shopping List
A user maintains a grocery list fpr monthly shopping Task:
Create a grocery list
Add new items to the list
Remove purchased items
Display the final shpping List'''

Grocerylist = ["Eggs", "Milk", " Berries", "Juices"]
# Add new items to the list
Grocerylist.append("Bread")
Grocerylist.append("Butter")
print(Grocerylist)
# Remove purchased items
Grocerylist.remove("Eggs")
Grocerylist.remove("Milk")
print(Grocerylist)
# Display the final shopping list
print("Final Shopping List:")
print(Grocerylist)
```

```python
# PROBLEM STATEMENT 4
''' 4. A person stores contacts in a list.
Taks:
Create a list of contacts
Add a new contact
Delete an existing contact
Display contatcs alphabetically'''

Contactlist = ["Jake", "Jason", "Aaron", "Joshua", "Aniket", "Guy"]

# Add a new contact
Contactlist.append("John")
print(Contactlist)

# Delete an existing contact
Contactlist.remove("Jake")
print(Contactlist)

# Display contacts alphabetically
Contactlist.sort()
print(Contactlist)
```

```python
# PROBLEM STATEMENT 5
''' 5. Temperature Analysis
Daily temperature readings of a city are stored in a list for 10 days.
Tasks:
Display temperature of first and last 5 days
Find highest and lowest temperature
Calculate average temperature'''

TempList = [22, 36, 12, 14, 32, 25, 21, 16, 26, 23]

# Display temperature of first and last 5 days
print("First 5 days temperature:", TempList[:5])
print("Last 5 days temperature:", TempList[-5:])

# Find highest and lowest temperature recorded
print("Highest temperature:", max(TempList))
print("Lowest temperature:", min(TempList))

# Calculate average temperature
print("Average temperature:", sum(TempList)/len(TempList))
```

```python
# PROBLEM STATEMENT 6
''' 6. Daily Expense Tracker
A person tracks daily expenses.
Tasks:
Store Daily expenses
Display total and average expenses
Identify highest spending day'''

# Store Daily Expenses
DailyExpense = [1000, 1500, 2000, 1500, 4000, 1000, 1000]

# Display Total and average expenses
print("Total expenses:", sum(DailyExpense))
print("Average expenses:", sum(DailyExpense)/len(DailyExpense))

# Identify highest spending day
print("Highest spending day:", DailyExpense.index(max(DailyExpense)))

# Identify lowest spending day
print("Lowest spending day:", DailyExpense.index(min(DailyExpense)))
```
